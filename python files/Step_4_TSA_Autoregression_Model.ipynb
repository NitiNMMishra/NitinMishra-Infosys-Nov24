# Autoregression Model

This notebook implements an Autoregression (AR) model for time series analysis and forecasting. The objective is to predict future values of a target variable based on its own past values.

The steps involved include:

- **Data Preprocessing**: Preparing the dataset for analysis.
- **Stationarity Testing**: Using the Dickey-Fuller test to check if the time series is stationary.
- **Model Fitting**: Applying the AR model to identify patterns in the data.
- **Forecasting**: Using the model to predict future values.
- **Evaluation**: Comparing the predicted values with the actual data to assess model performance.

The focus is on building a reliable model for accurate time series forecasting.


# Step 1: Importing Necessary Libraries

I began by importing the essential libraries to handle time series analysis, statistical operations, and visualizations effectively:

- **Pandas**: I am using it for data manipulation and analysis, such as cleaning and transforming the dataset.
- **Matplotlib**: I am leveraging this library to create visualizations that help me better understand the patterns in the time series data.
- **Statsmodels (AutoReg)**: I am utilizing the Autoregression (AR) model from `statsmodels` to forecast future values based on past data.
- **Statsmodels (ADF Test)**: The Augmented Dickey-Fuller test helps me assess whether the time series data is stationary or not, which is critical for time series modeling.


```python
# Step 1 : Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller
```

# Step 2: Loading and Preparing the Dataset

I began by loading the dataset and identifying the columns relevant for time series analysis:

- **Loading the Dataset**: 
  - I used the `pandas` library to read the Excel file containing the dataset. 
  - The dataset is stored in the variable `data` for further processing.
  - The file path `'../datasets/master_dataset/master_dataset.xlsx'` specifies the location of the dataset.

- **Identifying Numeric Time Series Columns**: 
  - I identified the key columns in the dataset that represent numeric time series data: 
    - `Clicks`: Represents the number of clicks over time.
    - `Impressions`: Represents the number of impressions over time.
    - `Quantity`: Represents the quantity values over time (the main column for forecasting).
  - These columns are stored in the list `time_series_columns` to streamline further operations.


```python
# Step 2: Load the dataset
data = pd.read_excel('../datasets/master_dataset/master_dataset.xlsx') 

# Step 3: Identify numeric time series columns
time_series_columns = ["Clicks", "Impressions", "Quantity"]  
```

# Steps 4 to 9: Performing Time Series Analysis (TSA)

In this section, I performed an end-to-end analysis of the numeric time series columns to uncover trends, fit a model, and make future predictions.

### Step 4: Looping Through Each Time Series Column
- I iterated over the list of numeric time series columns (`Clicks`, `Impressions`, `Quantity`) to analyze each column individually.
- For each column:
  - I removed missing values using the `.dropna()` method to ensure clean data for analysis.
  - I printed the first 5 rows of the time series to understand its structure.

### Step 5: Visualizing the Time Series
- I plotted the time series data for each column using `matplotlib`.
- The x-axis represents `Day Index` (time), and the y-axis represents the values of the respective column.
- This visualization helps me understand trends and patterns in the data.

### Step 6: Checking Stationarity with the Dickey-Fuller Test
- I used the **Augmented Dickey-Fuller (ADF) Test** to check whether the time series is stationary:
  - If the `p-value` is greater than 0.05, the data is non-stationary, and I applied **differencing** to make it stationary.
  - If the `p-value` is less than or equal to 0.05, the data is already stationary, and no differencing was applied.
- For non-stationary data:
  - I computed the differenced series using `.diff()`, which calculates the difference between consecutive values.

### Step 7: Fitting the Autoregression (AR) Model
- I fitted an **Autoregression (AR) Model** to the stationary time series:
  - The model was trained using a single lag (`lags=1`), which considers the immediate past value to predict the next value.
  - I printed the summary of the fitted model to evaluate its parameters and performance.

### Step 8: Making Predictions
- Using the fitted AR model, I predicted the values of the time series:
  - Predictions were generated for the entire dataset using the `.predict()` method.
  - I compared the **Actual** and **Predicted** values by combining them into a `pandas` DataFrame.
- I visualized the **Actual** vs. **Predicted** values to assess how well the model performed.

### Step 9: Forecasting Future Values
- I forecasted the next 5 values for the time series using the fitted AR model:
  - Forecasting helps me predict future trends based on past patterns.
- The forecasted values were displayed and can be used for decision-making or further analysis.

### Summary
Through this block, I analyzed the time series data column by column, ensured stationarity, fitted an AR model, and generated forecasts for future values. The visualizations and comparisons validate the accuracy of the model and its predictions.


```python
# Step 4: Looping through each column to conduct Time Series Analysis (TSA)
for column in time_series_columns:
    print(f"\n*** Analyzing column: {column}***")
    time_series = data[column].dropna()  # To remove the missing values

    # Display first few rows of the data
    print(f"\nFirst 5 rows of {column}:\n", time_series.head())

    # Step 5: Visualize the time series
    plt.figure(figsize=(10, 6))
    plt.plot(data["Day Index"], time_series, label=column, color="blue")
    plt.title(f"Time Series Data - {column}")
    plt.xlabel("Time")
    plt.ylabel(column)
    plt.legend()
    plt.grid()
    plt.show()

    # Step 6: Checking stationarity using the Dickey-Fuller test
    adf_test = adfuller(time_series)
    print(f"Dickey-Fuller Test Statistic: {adf_test[0]}")
    print(f"p-value: {adf_test[1]}")

    if adf_test[1] > 0.05:
        print(f"The column : {column} is non-stationary. Applying differencing...")
        time_series_diff = time_series.diff().dropna()  # Difference the series

        # Display first 5 rows of differenced series
        print(f"\nFirst 5 rows of differenced {column}:\n", time_series_diff.head())
    else:
        print(f"{column} is stationary.")
        time_series_diff = time_series

    # Step 7: Fit the Autoregression (AR) Model
    ar_model = AutoReg(time_series_diff, lags=1).fit()  # Default to a single lag

    # Print model summary
    print("AR Model Summary:")
    print(ar_model.summary())

    # Step 8: Make predictions
    predictions = ar_model.predict(start=1, end=len(time_series_diff) - 1)

    # Combine actual and predicted values for comparison
    comparison = pd.DataFrame({
        "Actual": time_series_diff[1:].values,
        "Predicted": predictions
    })

    # Display a few rows of the comparison
    print(f"\nFirst 5 rows of Actual vs Predicted for {column}:\n", comparison.head())

    # Plot Actual vs Predicted
    plt.figure(figsize=(10, 6))
    plt.plot(comparison["Actual"], label="Actual", color="blue")
    plt.plot(comparison["Predicted"], label="Predicted", color="red", linestyle="--")
    plt.title(f"Actual vs Predicted - AR Model ({column})")
    plt.xlabel("Time")
    plt.ylabel(column)
    plt.legend()
    plt.grid()
    plt.show()

    # Step 9: Forecast future values
    forecast_steps = 5  # Number of steps to forecast
    forecast = ar_model.predict(start=len(time_series_diff), end=len(time_series_diff) + forecast_steps - 1)
    print(f"\nForecasted Values for {column}:\n", forecast)
```
